```python
import pandas as pd
import requests
import numpy as np
from splinter import Browser
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
pd.options.mode.chained_assignment = None  # default='warn'
from time import sleep
from selenium.common.exceptions import NoSuchElementException
```

```python
# load in the dataframe of suggested movies from the csv
movies = pd.read_csv('suggested_movies.csv')
movies.head()
```

```python
movies[['title','imdb_id', 'URL', 'year', 'rating', 'runtime', 'imdb_rating']] = np.nan
```

```python
#check total amount of null values
movies.isnull().sum()
```

```python
#sort movies temporally
movies.sort_values(by=['week'], inplace=True)
```

```python
movies.head(20)
```

```python
#using selenium to get to web page of search query
driver = webdriver.Chrome()
driver.get('https://www.imdb.com/')
```

```python
def search_movie(movie):
    try:
        #check to see if there is pop up
        pop_up = driver.find_elements(By.XPATH, '/html/body/div[4]/div[2]/div/div[1]/button')
        if len(pop_up) > 0:
            pop_up[0].click()
            
        #input search
        search_elem = driver.find_element(By.ID, "suggestion-search")
        search_elem.clear()
        search_elem.send_keys(movie, Keys.RETURN)
        sleep(1)

        # navigate to first result
        driver.find_element_by_link_text("Movie").click()
        driver.find_element_by_xpath('//*[@id="main"]/div/div[2]/table/tbody/tr[1]/td[2]/a').click()
        sleep(1)
        # create soup env
        html = driver.page_source
        page_soup = soup(html,'html.parser')
        #locating metadata for movie
        url = driver.current_url
        url_list = url.split('/')
        imdb_id = url_list[4]


        results = page_soup.find('ul', class_='ipc-inline-list ipc-inline-list--show-dividers sc-8c396aa2-0 kqWovI baseAlt')
        data = results.find_all('li')

        #checking data len for non rated movies
        if len(data) == 3:
            # metadata
            year = data[0].a.text
            rating = data[1].a.text
            runtime = data[2].text
        elif len(data) == 2:
            # metadata
            year = data[0].a.text
            rating = 'NR'
            runtime = data[1].text

        # locating imdb rating
        results = page_soup.find('span', class_='sc-7ab21ed2-1 jGRxWM')
        imdb_rating = results.text

        # locating imdb title
        results = page_soup.find('h1')
        title = results.text

        # adding metadata to dataframe
        movies.loc[index, 'title'] = title
        movies.loc[index, 'imdb_id'] = imdb_id
        movies.loc[index, 'URL'] = url
        movies.loc[index, 'year'] = year
        movies.loc[index, 'rating'] = rating
        movies.loc[index, 'runtime'] = runtime
        movies.loc[index, 'imdb_rating'] = imdb_rating

        #printing results
        print(f'{movie} has been found and added to Dataframe.')
        sleep(2)
    except NoSuchElementException:
        print(f'ERROR: {movie} has not been found. Using Null Value')
        pass
```

```python
count = 0
for index, row in movies.iterrows():
#     if count == 15:
#         driver.quit()
#         break
    # taking break every 50th movie to not query site too quickly
    if count % 50 == 0 and count != 0:
            print('Break time...')
            sleep(30)
    query = row['movie_name']
    test = row['imdb_id']
    # check to see if the movie is a nan value
    if type(query) == float:
        count += 1
        continue
    
    print(f'Attempting to find: {query}...')
    # for rerunning loop if having an issue so not having to completely rerun program
    if type(test) == str:
        print(f'{query} has already been completed. Moving to next movie.')
        count += 1
        print(f'Movie {count} of {len(movies) + 1}')
        print('--------------------')
        continue
    
    search_movie(query)
    count += 1
    print(f'Movie {count} of {len(movies) + 1}')
    print('--------------------')
    
driver.quit()
```
