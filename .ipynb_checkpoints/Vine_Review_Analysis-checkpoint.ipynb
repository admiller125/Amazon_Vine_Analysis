```python
import pandas as pd
```

```python
file = "vine_table.csv" 
vine_data_df = pd.read_csv(file)
vine_data_df
```

```python
## get all reviews that have 20 or more total votes
total_20_df = vine_data_df.loc[vine_data_df["total_votes"] >= 20]
total_20_df
```

```python
## get all reviews that are in the program
helpful_50_df = vine_data_df.loc[vine_data_df["helpful_votes"]/vine_data_df["total_votes"] >= 0.5]
helpful_50_df
```

```python
## get all reviews that are in the vine program

vineview_y_df = vine_data_df.loc[vine_data_df["vine"] == "Y"]
vineview_y_df
```

```python
## get all reviews that are not in the vine program
vineview_n_df = vine_data_df.loc[vine_data_df["vine"] == "N"]
vineview_n_df
```

```python
total_reviews = vine_data_df["review_id"].count() 
five_star_reviews = vine_data_df.loc[vine_data_df["star_rating"] == 5].count()["review_id"]
five_star_vine = vine_data_df.loc[(vine_data_df["star_rating"] == 5) & (vine_data_df["vine"] == "Y")].count()["review_id"]
five_star_nonvine = vine_data_df.loc[(vine_data_df["star_rating"] == 5) & (vine_data_df["vine"] == "N")].count()["review_id"]
five_star_perc_vine = five_star_vine/vineview_y_df.count()["review_id"]*100
five_star_perc_nonvine = five_star_nonvine/vineview_n_df.count()["review_id"]*100
```

```python
print("The total number of reviews is",total_reviews)
print("The total number of 5 star reviews is",five_star_reviews)
print("The total number of 5 star vine reviews is",five_star_vine)
print("The total number of 5 star nonvine reviews is",five_star_nonvine)
print("The percentage of 5 star reviews in the vine program "+ str("{:.0f}".format(five_star_perc_vine)) +"%")
print("The percentage of 5 star reviews not in the vine program "+ str("{:.0f}".format(five_star_perc_nonvine)) +"%")
```
